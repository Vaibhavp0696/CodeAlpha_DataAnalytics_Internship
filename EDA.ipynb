### Task 2 : Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
df = sns.load_dataset("titanic")
```

```python
# Show first 5 rows of the dataset

df.head()
```

```python
# Overview of columns, types, and nulls
df.info()
```

```python
# Basic statistics for numerical columns
df.describe()
```

```python
# Count of null values in each column
df.isnull().sum()
```

```python
# Fill missing age values with mean
df["age"].fillna(df["age"].mean(), inplace=True)

# Drop rows with missing 'Embarked'
df.dropna(subset=["embarked"], inplace=True)

df.dropna(subset=["deck"], inplace=True)
```

```python
# Set theme
sns.set(style="whitegrid")
```

```python
# Survival count 

sns.countplot(x="survived", data = df)
plt.title("Survival Count")
plt.show()
```

```python
# survival by gender

sns.countplot(x="survived", hue="sex", data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
# Survival by Class
sns.countplot(x="survived", hue="class", data=df)
plt.title("Survival by Class")
plt.show()
```

```python
# Age distribution
sns.histplot(df["age"], bins = 30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# Age VS Class
sns.boxplot(x="class", y="age",data=df)
plt.title("Age by Class")
plt.show()
```

```python
sns.countplot(x="alive",hue="sex",data=df)
plt.title("Alive by Gender")
plt.show()
```

```python
# Average Fare by Survival Status
sns.barplot(x="survived", y="fare", data=df)
plt.title("Average Fare Paid by Survival Status")
plt.show()
```

```python
# Count of Male vs Female in Each Class
sns.countplot(x="class", hue="sex", data=df)
plt.title("Gender Distribution by Passenger Class")
plt.show()
```
